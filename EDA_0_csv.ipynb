```python
import pandas as pd 
df = pd.DataFrame()
```

# to csv

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
temp_df=df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_csv('batsman_runs.csv',index = False)
```

```python
df.pivot_table(index = 'batsman', columns = 'bowling_team', values = 'batsman_runs', aggfunc = 'sum').to_csv('batsman pivot.csv' )
```

# to excel

```python
temp_df=df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_excel('batsman_runs.xlsx')
```

```python
temp_df.to_excel("output.xlsx",sheet_name = 'batmans_runs')
```

```python
temp_df2 = df.pivot_table(index = 'batsman', columns = 'bowling_team', values = 'batsman_runs', aggfunc = 'sum')
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    temp_df.to_excel(writer, sheet_name = 'sheet_name1')
    temp_df2.to_excel(writer, sheet_name = 'sheet_name2')
```

# to html


```python
df.query('batsman_runs == 6').pivot_table(index = 'over', columns = 'ball', values = 'batsman_runs', aggfunc = 'count').to_html('sixes_hitmap.html')
```

# Gather data through API or Web Scraping

```python
import pandas as pd
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
```

```python
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
```

```python
df = pd.DataFrame()
```

```python
dfs = []

for i in range(1, 429):
    response = requests.get(
        f"https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}"
    )
    temp_df = pd.DataFrame(response.json()['results'])[
        ['id','title','overview','release_date','popularity','vote_average','vote_count']
    ]
    dfs.append(temp_df)

df = pd.concat(dfs, ignore_index=True)
```
